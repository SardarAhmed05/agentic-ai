```python
# Sync
import requests
import time

def get_weather(city):
    response = requests.get(f"https://api.open-meteo.com/v1/forecast?latitude=24.86&longitude=67.00&current_weather=true")
    return response.json()["current_weather"]["temperature"]

start = time.time()
get_weather("Karachi")
get_weather("Lahore")
get_weather("Islamabad")
print(f"Time: {time.time() - start:.2f}s")
```

```python
# Async
import asyncio
import aiohttp
import time

async def get_weather(session, city, lat, lon):
    url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&current_weather=true"
    async with session.get(url) as response:
        data = await response.json()
        return data["current_weather"]["temperature"]

async def main():
    async with aiohttp.ClientSession() as session:
        results = await asyncio.gather(
            get_weather(session, "Karachi", 24.86, 67.00),
            get_weather(session, "Lahore", 31.55, 74.34),
            get_weather(session, "Islamabad", 33.68, 73.04)
        )
        print(results)

start = time.time()
await main()
print(f"Time: {time.time() - start:.2f}s")
```

```python
# Runs both functions simultaneously, so the total time taken is 2 seconds instead of 4 seconds.

import asyncio
import time

async def greet(name, delay):
    await asyncio.sleep(delay)
    print(f"Hello {name}")

async def main():
    start = time.time()
    await asyncio.gather(
        greet("Sardar", 2),
        greet("Ahmed", 2)
    )
    print(f"Time: {time.time() - start:.2f}s")

await main()
```

```python
# aiohttp is the async version of requests library. It is used to make asynchronous HTTP requests in Python.
import aiohttp

async def fetch(url):
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as response:
            data = await response.json()
            return data

result = await fetch("https://api.open-meteo.com/v1/forecast?latitude=24.86&longitude=67.00&current_weather=true")
print(result)
```
